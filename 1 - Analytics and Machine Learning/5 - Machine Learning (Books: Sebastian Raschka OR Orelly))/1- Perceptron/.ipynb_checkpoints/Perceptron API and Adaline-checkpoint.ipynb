# Implementação do algoritmo Perceptron do livro "Python Machine Learning - Sebastian Raschka" <br>

Inicalizaremos o objeto _Perceptron_ que irá aprender com os dados através do metódo "fit", e fará previsões pelo metódo "predict" <br>

___OBS:Por convenção atributos de classes que possuirem um underscore serão atributos que chamaram outros metodos do objeto___

```python
import numpy as np
```

```python
class Perceptron(object):
    """Perceptron Classificador
    
    Parametros
    ------------
    eta: float
        Taxa de Aprendizagem (varia entre 0.0 e 1.0)
    n_iter: int
        Passos sobre o dataset de treinamento (epochs)
    random_state: int
        Numero aleatório gerador dos seeds para inicialização com pesos randomicos
    
    Atributos
    -----------
    w_: Vetor 1D
        Pesos depois do fitting
    errors_: list
        Numeros de Classificações erradas em cada epoch
        
    """
    
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        
    def fit(self, X, y):
        """Fit dos dados de treino
        
        Parametros
        ------------
        X: vetor, shape = [n_samples, n_features]
            Vetores de treinamento, onde n_samples é o numero de amostras e
            n_features o numero de features.
        y: vetor, shape = [n_samples]
            Valores targets
            
        Returns
        ---------
        self: object
        
        """
        rgen = np.random.RandomState(self.random_state) #Gera numeros aleatórios de uma variedade de distribuições de probabilidade
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1]) # .normal amostra de distribuição gaussiana normal (normalização)
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self
    
    def net_input(self, X):
        "Calculando o net input"
        return np.dot(X, self.w_[1:]) + self.w_[0] #np.dot multipla os vetores de forma cruzada, ou seja,
                                                          #é o mesmo que fazer W(transposta) * X
    def predict(self, X):
        "Retorna a label da classe"
        return np.where(self.net_input(X) >= 0.0, 1, -1)
    
```

## Treinando o Perceptron com o dataset Iris

___Irei usar apenas duas classes e duas features do dataset iris, apenas por praticidade, porem o algoritmo pode ser usado não esta restrito a apenas duas dimensões de dados___

```python
import pandas as pd
df = pd.read_csv('/home/leandro/Documents/Data/iris.data', header=None)
df.head()
```

Irei extrair 100 classes do dataset, sendo 50 da classe Iris-setosa e 50 da classe Iris-versicolor, e converter as labels dessas classes para dois inteiros, label 1 (versicolor) e label -1 (setosa) que serão passadas ao vetor y.<br>
De modo análogo extrairei a primeira coluna (feature) do dataset - comprimento da sepal-  e a terceira coluna (comprimento da petala) das 100 amostras de treino e atribui-las a matrix X para entrão plota-las através de um scatter plot.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
#Selecionando setosa e versicolor
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)

#Selecionando comprimento de sepal e comprimento da petala
X = df.iloc[0:100, [0,2]].values
```

```python
#Plot data
plt.figure(figsize=(14,9))
plt.scatter(X[:50, 0], X[:50, 1],
           color='red', marker='o', label='setosa')

plt.scatter(X[50:100, 0], X[50:100, 1],
           color='blue', marker='x', label='versicolor')

plt.xlabel('Comprimento Sepal [cm]')
plt.ylabel('Comprimento Pétala [cm]')
plt.legend(loc='upper left')
plt.show()
```

__Podemos notar que uma fronteira linear de decisão é suficiente para parar as flores Setosas das Versicolors. Portanto, um classificador linear, como o perceptron deve ser capaz de classificar as flores deste dataset__

## Treinando o Perceptron

_Além de extrair um sub-dataset para treino, irei plotar os erros de classificação para cada epoch para checar se o algoritmo convergiu e encontrou a fronteira de decisão que separa as duas classes de flores_

```python
#Instanciando um objeto da classe Perceptron
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X, y)

plt.figure(figsize=(14,9))
plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Numero de updates')
plt.show()
```

__Vemos que nosso perceptron converge depois de 6 epocas e passa a classificar corretamente as amostras de treino__<br><br>
Vamos implementar uma função para visualizar, de forma mais conveniente, a fronteira de decisão para nossos dados

```python
from matplotlib.colors import ListedColormap
#Um objeto da classe Colormap é gerado apartir de uma lista de cores
```

```python
def plot_decision_regions(X, y, classifier, resolution=0.02):
    #Setup do gerador de marcadores e cores (color map)
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    #plot da superficie de decisão
    #definindo o min e max das duas features
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    #Usando o min e max das features para criar um par de vetores de grid para a função meshgrid
    #np.meshgrid cria uma grade retangular apartir de duas matrizes para uma representação cartesiana
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    """
    _Como treinamos nosso perceptron com dados de duas diemensões, precisamos "achatar" os vetores grid,
    para criar uma matrix que tenha o mesmo numero de colunas que o sub-dataset de treino,
    então usamos o metodo predict para prever as lebels Z correpondentes aos pontos do grid.
    _.ravel() retorna um vetor "achatado", em 1D
    _.T acessa o atributo 'T' do objeto, que Transpoẽ o vetor (vetor transposto)
    _.shape() retorna uma tupla com as dimensões do vetor
    
    """
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    #Após redimensionar Z dentro do grid para as mesmas dimensões de xx1 e xx2, podemos então
    #plotar o seu contorno, onde diferentes regiões de decisão terão cores diferentes
    plt.figure(figsize=(14,9))
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    #Plot das classes das amostras
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```

```python
plot_decision_regions(X, y, classifier=ppn)

#plt.figure(figsize=(14,9))
plt.xlabel('Comprimento Sepal [cm]')
plt.ylabel('Comprimento da Pétala [cm]')
plt.legend(loc='upper left')
plt.show()
```

___Como vemos no plot, o Perceptron aprendeu uma fronteira de decisão limite capaz de classificar corretamente as flores da nossa amostra___
